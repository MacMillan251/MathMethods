Задание 1. Модель Леонтьева

```python
import numpy as np


input_matrix = [[6, 25], [15, 20]]
consume_vector = [90, 130]
production_vector = [121, 165]
new_vector = [180, 120]

X = np.array(input_matrix)
y = np.array(consume_vector)
x = np.array(production_vector)
y2 = np.array(new_vector)

# Единичная матрица 2х2
E = np.eye(2)
    
# Построение матрицы прямых затрат A
A = np.zeros((2, 2))
A[0, 0] = X[0, 0] / x[0]
A[1, 0] = X[1, 0] / x[0]
A[0, 1] = X[0, 1] / x[1]
A[1, 1] = X[1, 1] / x[1]

# Построение матрицы E-A
EA = E - A
print("EA = ", EA, "\n")

# Построение матрицы полных затрат S=(E-A)^-1
S = np.linalg.inv(EA)
print("S = ", S, "\n")

# Проверка обратной матрицы
E1 = EA * S
print("E1 = ", E1, "\n")

# Проверка каждого элемента матрицы S на положительность
matrixisproductive = (S>0).all()
print("Матрица продуктивна") if matrixisproductive else print("Матрица не продуктивна")

# Вычисление собственных значений и векторов матрицы A
values, vectors = np.linalg.eig(A)
print("Values = ", values, "\n")
print("Vectors = ", vectors, "\n")

# Число Фробениуса Альфа
lambda_A = max(values)
print("Lambda A = ", lambda_A, "\n")

# Векторы A, стоящие в столбцах матрицы P
# Вектор Фробениуса 
P = vectors

# Вектор объемов производства по отраслям (x1=x)
x1 = np.matmul(S, y)
print("x1 = ", x1, "\n")

# Вектор новых объемов производства по отраслям
x2 = np.matmul(S, y2)
print("x2 = ", x2, "\n")

xc = np.array([0, 0])
x11 = A[0, 0] * x2[0]
x21 = A[1, 0] * x2[0]
    
xc[0] = x2[0] - (x11+x21)
x12 = A[0, 1] * x2[1]
x22 = A[1, 1] * x2[1]
xc[1] = x2[1] - (x12+x22)

print("xc = ", xc, "\n")
```
